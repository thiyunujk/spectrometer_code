```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate the noisy bell curve with a specified center index position
def generate_noisy_curve(center_index, num_elements):
    # Generate a smooth curve with multiple peaks using a Gaussian distribution
    x = np.linspace(-3, 3, num_elements)
    gaussian_values = np.exp(-x ** 2)  # Adjust sigma/variance for wider/narrower peaks

    # Scale the Gaussian values to fit within the desired range
    gaussian_values *= 900  # Adjust the height of the curve

    # Store the peak value at the specified center index
    center_peak = gaussian_values[center_index]

    # Add noise to the curve without altering the center peak
    noise = np.random.normal(0, 50, num_elements)  # Adjust the noise intensity
    noisy_curve = gaussian_values + noise - (gaussian_values[center_index] - center_peak)

    return noisy_curve

# Define the number of elements
num_elements = 418

# Initial center index
center_index = num_elements // 2

# Generate the noisy curve with the initial center index
noisy_curve = generate_noisy_curve(center_index, num_elements)

# Plot the initial noisy curve
plt.figure(figsize=(8, 6))
plt.plot(range(num_elements), noisy_curve)
plt.title(f'Noisy Bell Curve-Like Graph with Center at Index {center_index}')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Change the center index position (for example, moving it 50 positions to the right)
new_center_index = center_index + 200

# Generate the noisy curve with the new center index
noisy_curve_new_center = generate_noisy_curve(new_center_index, num_elements)

# Plot the noisy curve with the new center index
plt.figure(figsize=(8, 6))
plt.plot(range(num_elements), noisy_curve_new_center)
plt.title(f'Noisy Bell Curve-Like Graph with Center at Index {new_center_index}')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
