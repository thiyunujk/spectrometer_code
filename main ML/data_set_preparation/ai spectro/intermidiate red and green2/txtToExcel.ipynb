```python
import numpy as np
import openpyxl

# Read the Excel file
wb = openpyxl.load_workbook('Reflection_20-31-01-730.xlsx')
ws = wb.active

# Extract the first column data and skip the first row
data = []
for row_index, row in enumerate(ws.iter_rows()):
    if row_index == 0:  # Skip the header row
        continue

    cell = row[0]
    if cell.value is not None:
        data.append(float(cell.value))

# Convert the data to a NumPy array
data_array = np.array(data)

# Convert the NumPy array to a comma-separated list
csv_list = ','.join(str(item) for item in data_array.tolist())

# Add the text ",raw" to the end of the CSV list
csv_list += ",raw"

# Create a new Excel file
new_wb = openpyxl.Workbook()
new_ws = new_wb.active

# Write the comma-separated list to the first cell
new_ws.append(['csv_list'])
new_ws['A1'] = csv_list

# Save the new Excel file
new_wb.save('new_file.xlsx')
```
